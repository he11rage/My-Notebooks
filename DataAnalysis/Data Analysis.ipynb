```python
import pandas as pd
```

```python
df = pd.read_csv("wage.csv")
df
```

```python
df.gender = df.gender.apply(lambda g: "M" if g == 0 else "F")
```

```python
df.T
```

```python
df.groupby("gender")["wage"].mean()
```

```python
df.value_counts()
```

```python
df.nunique()
```

```python
df = df.drop_duplicates()
df
```

```python
df.wage.describe()
```

```python
df = df[df.wage > 0]
df.wage.describe()
```

```python
bonus = pd.read_csv("bonus.csv", sep=";")
bonus
```

```python
jf = pd.merge(df, bonus, how = "outer")
jf
```

```python
jf.bonus = jf.bonus.apply(lambda b: 0 if pd.isna(b) else b)
jf
```

```python
jf["total"] = jf.wage * 12 + jf.bonus
jf
```

```python
jf.groupby("gender")["total"].agg(["mean", "median"])
```

```python
jf.to_csv("new_wage.csv")
```
